# Ex - 4

```python
import xml.etree.ElementTree as ET
tree = ET.parse('Data/breakfast_menu.xml')
root = tree.getroot()
data = []
for food in root.findall("food"):
    item = {
        "name": food.find("name").text,
        "price": food.find("price").text,
        "description": food.find("description").text,
        "calories": int(food.find("calories").text)
    }
    data.append(item)

df = pd.DataFrame(data)

df
```

```python
df.to_csv('Data/breakfast_menu.csv')
```

# Ex - 7

```python
temp_df = pd.read_csv('Data/nhtemp.csv')
temp_df
```

```python
temp_df = temp_df.iloc[:,1:]
temp_df
```

```python
sns.distplot(temp_df["nhtemp"], hist = True, kde = False)
plt.title("Histogram of nhtemp")
plt.xlabel("nhtemp")
plt.ylabel("Frequency")
plt.show()
```

```python
temp_df.nhtemp.describe(percentiles=[0.25, 0.75])
```

```python
five_point_summary = temp_df['nhtemp'].describe()[['min', '25%', '50%', '75%', 'max']]
five_point_summary
```

```python
# Identify outliers based on the IQR method
Q1 = temp_df['nhtemp'].quantile(0.25)
Q3 = temp_df['nhtemp'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Detecting outliers
outliers = temp_df[(temp_df['nhtemp'] < lower_bound) | (temp_df['nhtemp'] > upper_bound)]
outlier_count = outliers.shape[0]
outlier_count
```

```python
outliers
```

```python
sns.boxplot(y=temp_df["nhtemp"])
plt.title("Boxplot of nhtemp")
plt.show()
```
