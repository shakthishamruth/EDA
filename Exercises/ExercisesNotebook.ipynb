# Ex - 4

```python
import xml.etree.ElementTree as ET
tree = ET.parse('Data/breakfast_menu.xml')
root = tree.getroot()
data = []
for food in root.findall("food"):
    item = {
        "name": food.find("name").text,
        "price": food.find("price").text,
        "description": food.find("description").text,
        "calories": int(food.find("calories").text)
    }
    data.append(item)

df = pd.DataFrame(data)

df
```

```python
df.to_csv('Data/breakfast_menu.csv')
```

```python
dataAir = pd.read_csv('Data/AirPassengers.csv')
first_half_df = dataAir.iloc[:,:6]
second_half_df = dataAir.iloc[:,6:]
```

```python
first_half_df.head()
```

```python
second_half_df.head()
```

```python
air_df = first_half_df.merge(second_half_df, left_index=True, right_index=True, how='inner')
air_df
```

```python
air_df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
df = air_df.loc[:,["Feb"]]
df[df.isna().any(axis=1)].index.tolist()
```

```python
air_df['Feb'] = air_df['Feb'].fillna(air_df['Feb'].median())
air_df['Mar'] = air_df['Mar'].fillna(air_df['Mar'].median())
air_df['Jun'] = air_df['Jun'].fillna(air_df['Jun'].median())
air_df['Aug'] = air_df['Aug'].fillna(air_df['Aug'].median())
air_df['Oct'] = air_df['Oct'].fillna(air_df['Oct'].median())
air_df['Nov'] = air_df['Nov'].fillna(air_df['Nov'].median())
air_df['Dec'] = air_df['Dec'].fillna(air_df['Dec'].median())
air_df
```

```python
air_df.isna().sum()
```

```python
air_df['Year'] = pd.Series(range(1949, 1961))
air_df
```
